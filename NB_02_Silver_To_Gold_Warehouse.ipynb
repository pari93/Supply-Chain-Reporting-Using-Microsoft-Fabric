```sql
-- Run this as a T-SQL notebook attached to WH_Evos_Gold.

MERGE gold.dim_terminal AS t
USING [LH_Evos_BronzeSilver].[dbo].[silver_terminal] AS s
ON t.terminal_code = s.terminal_code
WHEN MATCHED AND s.last_modified_utc >= t.last_modified_utc THEN UPDATE SET
    terminal_name = s.terminal_name,
    country = s.country,
    port_hub = s.port_hub,
    total_capacity_cbm = s.total_capacity_cbm,
    number_of_tanks = s.number_of_tanks,
    operational_since = s.operational_since,
    specialty = s.specialty,
    source_url = s.source_url,
    last_modified_utc = s.last_modified_utc,
    dw_loaded_utc = SYSUTCDATETIME()
WHEN NOT MATCHED THEN INSERT (
    terminal_code, terminal_name, country, port_hub, total_capacity_cbm, number_of_tanks,
    operational_since, specialty, source_url, last_modified_utc, dw_loaded_utc
) VALUES (
    s.terminal_code, s.terminal_name, s.country, s.port_hub, s.total_capacity_cbm, s.number_of_tanks,
    s.operational_since, s.specialty, s.source_url, s.last_modified_utc, SYSUTCDATETIME()
);

MERGE gold.dim_product AS t
USING [LH_Evos_BronzeSilver].[dbo].[silver_product] AS s
ON t.product_code = s.product_code
WHEN MATCHED AND s.last_modified_utc >= t.last_modified_utc THEN UPDATE SET
    product_name = s.product_name,
    product_family = s.product_family,
    is_new_energy = s.is_new_energy,
    typical_density_kg_m3 = s.typical_density_kg_m3,
    last_modified_utc = s.last_modified_utc,
    dw_loaded_utc = SYSUTCDATETIME()
WHEN NOT MATCHED THEN INSERT (
    product_code, product_name, product_family, is_new_energy, typical_density_kg_m3, last_modified_utc, dw_loaded_utc
) VALUES (
    s.product_code, s.product_name, s.product_family, s.is_new_energy, s.typical_density_kg_m3, s.last_modified_utc, SYSUTCDATETIME()
);

MERGE gold.dim_customer_contract AS t
USING [LH_Evos_BronzeSilver].[dbo].[silver_contract] AS s
ON t.contract_id = s.contract_id
WHEN MATCHED AND s.last_modified_utc >= t.last_modified_utc THEN UPDATE SET
    customer_code = s.customer_code,
    customer_name = s.customer_name,
    customer_type = s.customer_type,
    customer_country = s.customer_country,
    terminal_code = s.terminal_code,
    product_code = s.product_code,
    contract_start_date = s.contract_start_date,
    contract_end_date = s.contract_end_date,
    reserved_capacity_cbm = s.reserved_capacity_cbm,
    storage_fee_eur_per_cbm_month = s.storage_fee_eur_per_cbm_month,
    handling_fee_eur_per_cbm = s.handling_fee_eur_per_cbm,
    contract_status = s.contract_status,
    last_modified_utc = s.last_modified_utc,
    dw_loaded_utc = SYSUTCDATETIME()
WHEN NOT MATCHED THEN INSERT (
    contract_id, customer_code, customer_name, customer_type, customer_country, terminal_code, product_code,
    contract_start_date, contract_end_date, reserved_capacity_cbm, storage_fee_eur_per_cbm_month,
    handling_fee_eur_per_cbm, contract_status, last_modified_utc, dw_loaded_utc
) VALUES (
    s.contract_id, s.customer_code, s.customer_name, s.customer_type, s.customer_country, s.terminal_code, s.product_code,
    s.contract_start_date, s.contract_end_date, s.reserved_capacity_cbm, s.storage_fee_eur_per_cbm_month,
    s.handling_fee_eur_per_cbm, s.contract_status, s.last_modified_utc, SYSUTCDATETIME()
);

MERGE gold.dim_date AS t
USING [LH_Evos_BronzeSilver].[dbo].[silver_date] AS s
ON t.date_key = s.date_key
WHEN MATCHED THEN UPDATE SET
    year_number = s.year_number,
    month_number = s.month_number,
    month_name = s.month_name,
    day_of_month = s.day_of_month,
    iso_week = s.iso_week,
    day_of_week_number = s.day_of_week_number,
    dw_loaded_utc = SYSUTCDATETIME()
WHEN NOT MATCHED THEN INSERT (
    date_key, year_number, month_number, month_name, day_of_month, iso_week, day_of_week_number, dw_loaded_utc
) VALUES (
    s.date_key, s.year_number, s.month_number, s.month_name, s.day_of_month, s.iso_week, s.day_of_week_number, SYSUTCDATETIME()
);

MERGE gold.fact_terminal_movement AS t
USING [LH_Evos_BronzeSilver].[dbo].[silver_movement] AS s
ON t.movement_id = s.movement_id
WHEN MATCHED AND s.last_modified_utc >= t.last_modified_utc THEN UPDATE SET
    movement_datetime_utc = s.movement_datetime_utc,
    planned_datetime_utc = s.planned_datetime_utc,
    terminal_code = s.terminal_code,
    product_code = s.product_code,
    contract_id = s.contract_id,
    movement_type = s.movement_type,
    transport_mode = s.transport_mode,
    volume_cbm = s.volume_cbm,
    movement_status = s.movement_status,
    delay_hours = s.delay_hours,
    delay_reason = s.delay_reason,
    last_modified_utc = s.last_modified_utc,
    movement_date = s.movement_date,
    is_completed = s.is_completed,
    is_delayed = s.is_delayed,
    estimated_handling_fee_eur = s.estimated_handling_fee_eur,
    dw_loaded_utc = SYSUTCDATETIME()
WHEN NOT MATCHED THEN INSERT (
    movement_id, movement_datetime_utc, planned_datetime_utc, terminal_code, product_code, contract_id,
    movement_type, transport_mode, volume_cbm, movement_status, delay_hours, delay_reason,
    last_modified_utc, movement_date, is_completed, is_delayed, estimated_handling_fee_eur, dw_loaded_utc
) VALUES (
    s.movement_id, s.movement_datetime_utc, s.planned_datetime_utc, s.terminal_code, s.product_code, s.contract_id,
    s.movement_type, s.transport_mode, s.volume_cbm, s.movement_status, s.delay_hours, s.delay_reason,
    s.last_modified_utc, s.movement_date, s.is_completed, s.is_delayed, s.estimated_handling_fee_eur, SYSUTCDATETIME()
);

SELECT 'Gold load completed' AS status_message;
```
